# Area Plot/Chart

```python
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='skyblue',alpha=0.4)
plt.plot(x,y,color='slateblue',alpha=0.8)#for border of area plot
plt.show()
```

```python
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```

# Word Cloud

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
alice_novel=open('alice_in_wonderland.txt')
alice_novel=alice_novel.read()
stopwords=set(STOPWORDS)
print(stopwords)
print(len(stopwords))
```

```python
alice_wc=WordCloud(background_color="white",max_words=2000,stopwords=stopwords,collocations=False)
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
df_incidents=pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_incidents.head()
```

```python
df=df_incidents.iloc[0:100]
df.shape
```

```python
import folium
latitude=37.77
longitude=-122.42
sanfran_map=folium.Map(location=[latitude,longitude],zoom_start=12)
incidents=folium.map.FeatureGroup()
for lat,lon in zip(df.Y,df.X):
    incidents.add_child(folium.CircleMarker([lat,lon],radius=5,color='yellow',fill=True,fill_color='black',fill_opacity=0.6))
sanfran_map.add_child(incidents)
```

```python
import pandas as pd
state_unemp=pd.read_csv("US_Unemployment_Oct2012.csv")
state_geo='us-states.json'
```

```python
state_unemp.head()
```

```python
import folium
usa_state=folium.Map(location=[48,-102],zoom_start=3)
folium.Choropleth(geo_data=state_geo,name='choropleth',data=state_unemp,columns=['State','Unemployment'],key_on='feature.id',fill_color='YlGnBu',fill_opactiy=0.7,line_opacity=0.5,legend_name='Unemployment Scale').add_to(usa_state)
usa_state
```

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4,5,6])
G.add_edge(1,2)
G.add_edge(1,1)
G.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(G,node_size=850,node_color='red',width=2,edge_color='blue')
plt.show()
```
